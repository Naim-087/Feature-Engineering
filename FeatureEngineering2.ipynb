#**Label encoding**

## import

```python
from sklearn.preprocessing import LabelEncoder
```

## label cols

```python
label_cols =['aspiration','enginelocation']

le=LabelEncoder()
```

```python
for col in label_cols:
  df[col] = le.fit_transform(df[col])
```

## View LabelEncoded cols

```python
df[['aspiration','enginelocation']]
```

```python
# View unique encoded values

print(df['aspiration'].unique())

print(df['enginelocation'].unique())
```

# **Ordinal Encoding**

## import

```python
from sklearn.preprocessing import OrdinalEncoder
```

## ordinal cols

```python
ordinal_cols =['doornumber','cylindernumber']

ord_encoder =OrdinalEncoder( categories=[
        ['two', 'four'],
        ['two', 'three', 'four', 'five', 'six', 'eight', 'twelve']
    ])
```

## apply ordinal encoding

```python
df[ordinal_cols]=ord_encoder.fit_transform(df[ordinal_cols])
```

## view encoded cols

```python
df[['doornumber','cylindernumber']].head(15)
```

# One-hot-encoding

## import

```python
from sklearn.preprocessing import OneHotEncoder
```

## one-hot-cols

```python
onehot_cols = [
    'fueltype',
    'carbody',
    'drivewheel',
    'enginetype',
    'fuelsystem'
]
```

image.png

```python
df = pd.get_dummies(
    df,
    columns=onehot_cols,
    drop_first=True
)
```

```python
# ==============================
# VIEW ONE HOT ENCODED COLUMNS
# ==============================

onehot_cols = [col for col in df.columns if '_' in col]

print(df[onehot_cols].head())
```
